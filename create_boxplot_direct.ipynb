```python
import numpy as np
import matplotlib.pyplot as plt
from tifffile import tifffile as tiff
import pandas as pd
from itertools import zip_longest

print("running")
loc = ("D:\\A_MN_figures\\pdgfb2_photonsbinned\\")


mid = np.array(tiff.imread(f'{loc}CROPPED_pdgfb2_mid.tif'),dtype=float)
cortex = np.array(tiff.imread(f'{loc}CROPPED_pdgfb2_cortex.tif'),dtype=float)
cerebellum =  np.array(tiff.imread(f'{loc}CROPPED_pdgfb2_cerebellum.tif'),dtype=float)

data = {'mid': mid[mid>0],
        'cortex': cortex[cortex>0],
        'cerebellum': cerebellum[cerebellum>0]}
#TODO switch to numpypad 
zipped_data = (zip_longest(*data.values(), fillvalue=np.nan))

df = pd.DataFrame(zipped_data, columns=['mid', 'cortex','cerebellum'])
#boxplot = df.boxplot(grid=False, fontsize=15)

print(df.mid)
```
